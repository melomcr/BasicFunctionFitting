```python
import numpy as np
from scipy.optimize import curve_fit as cfit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(ang, k, n, d):
    return k * (1 + np.cos(n*ang - d))
```

```python
xdata = (np.linspace(0, 360, 13))*np.pi/180
energies = np.asarray([0,-4.37,-7.06,-3.69,.00,-4.32,-7.01,-3.74,-0.01,-4.44,-7.01,-3.69,0])

plt.plot(xdata, energies, 'ro', label='data')
```

```python
popt, pcov = cfit(func, xdata, energies, p0=[-3,3,3])

plt.plot(xdata, func(xdata, *popt), 'b-o',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.plot(xdata, energies, 'ro', label='data')
plt.legend()
```
